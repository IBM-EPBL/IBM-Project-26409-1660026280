```python
q1=ds.Age.quantile(0.25) #q1
q3=ds.Age.quantile(0.75) #q2
```

```python
IQR=q3-q1
```

```python
upper_limit=q3+1.5*IQR
lower_limit=q1+1.5*IQR
```

```python
ds=ds[ds.Age<upper_limit]
```
